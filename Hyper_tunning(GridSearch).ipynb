```python
from sklearn import datasets

Data = datasets.load_iris()
print(dir(Data))
Data.target_names
```

```python
import pandas as pd

File = pd.DataFrame(Data.data,columns=Data.feature_names)
File['target'] = Data.target
File['target_name'] = File['target'].apply(lambda x : Data.target_names[x])
X = Data.data
Y = Data.target
File[40:60]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,Y,test_size=0.3)
y_train
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='rbf',C=1)
svc.fit(X_train,y_train)
svc.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy

print(cross_val_score(SVC(kernel='poly',C=2,gamma='auto'),X,Y,cv=4))
print(cross_val_score(SVC(kernel='rbf',C=5,gamma='auto'),X,Y,cv=4))
print(cross_val_score(SVC(kernel='linear',C=5,gamma='auto'),X,Y,cv=4))
```

```python
kernel = ['poly','linear','rbf','sigmoid']
C = [10,20,30]
Avg_score = {}
for kval in kernel:
    for cal in C:
        cv_score = cross_val_score(SVC(kernel=kval,C=cal,gamma='auto'),X,Y,cv=4)
        Avg_score[kval + '_'+ str(cal)] = numpy.average(cv_score),    

Avg_score
```

```python
from sklearn.model_selection import GridSearchCV

Grid_model = GridSearchCV(SVC(gamma='auto'), 
                              {   'kernel':['linear','poly','rbf'],
                                  'C':[10,20,13,23] 
                              },cv=4,return_train_score=False)
Grid_model.fit(X,Y)
Grid_model.cv_results_
```

```python
import pandas
data = pandas.DataFrame(Grid_model.cv_results_)
data[['rank_test_score','params','param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV

Random_Cv = RandomizedSearchCV(SVC(gamma='auto'),
                               {
                                   'kernel' :['linear','poly','sigmoid'],
                                   'C' : [10,20,30]    
                               },
                               cv=4,
                               return_train_score=False
                               )
Random_Cv.fit(X,Y)
data = pd.DataFrame(Random_Cv.cv_results_)
data[['rank_test_score','params','param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import linear_model

Models = {
    'Random_Forest':{
                                'model': RandomForestClassifier(),
                                'param':{
                                            'n_estimators':[1,10,5],
                                            'n_jobs':[2,5,6]
                                         }
                            },
            'Log_Regress':{
                                'model':LogisticRegression(solver='liblinear',multi_class='auto'),
                                'param':{
                                            'C':[1,3]
                                        }
                          },
            'svm':{
                    'model': SVC(gamma='auto'),
                    'param':{
                        'kernel':['linear','rbf'],
                        'C':[1,3,7,4]
                            }
                  }
            
         }

score = []

for model_name,mp in Models.items():
    Grid_model = GridSearchCV(mp['model'],mp['param'],cv=3,return_train_score=False)
    Grid_model.fit(X,Y)
    score.append({'model':model_name,
                  'best_score':Grid_model.best_score_,
                  'best_Param':Grid_model.best_params_,
                  'index':Grid_model.best_index_
                  })
Data = pd.DataFrame(score,columns=['model','best_score','best_Param','index'])
Data
```
