```python
from sklearn import datasets
import pandas 
Data = datasets.load_iris()
print(dir(Data))
File = pandas.DataFrame(Data.data,columns=Data.feature_names)
File['Target'] = Data.target
File['Flower_Name'] = File['Target'].apply(lambda x:Data.target_names[x])
File
```

```python
from sklearn.cluster import KMeans
Km = KMeans(n_clusters=3)
Predict = Km.fit_predict(File[['petal length (cm)','petal width (cm)']])
File['Cluster'] = Predict
File
```

```python
import matplotlib.pyplot as plt

File1 = File[File['Target']==0]
File2 = File[File['Target']==1]
File3 = File[File['Target']==2]

plt.scatter(File1['petal width (cm)'],File1['petal length (cm)'],color='red')
plt.scatter(File2['petal width (cm)'],File2['petal length (cm)'],color='green')
plt.scatter(File3['petal width (cm)'],File3['petal length (cm)'],color='blue')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

Scale = MinMaxScaler()
Scale.fit(File[['petal width (cm)']])
File['petal width (cm)'] = Scale.transform(File[['petal width (cm)']])

Scale.fit(File[['petal length (cm)']])
File['petal length (cm)'] = Scale.transform(File[['petal length (cm)']])
File
```

```python
Km = KMeans(n_clusters=3)
Scale_Predict =Km.fit_predict(File[['petal length (cm)','petal width (cm)']])
File['Scale_Predict'] = Scale_Predict
File
```

```python
File1 = File[File['Scale_Predict']==0]
File2 = File[File['Scale_Predict']==1]
File3 = File[File['Scale_Predict']==2]

plt.scatter(File1['petal length (cm)'],File1['petal width (cm)'],color='red',label='red')
plt.scatter(File2['petal length (cm)'],File2['petal width (cm)'],color='green',label='green')
plt.scatter(File3['petal length (cm)'],File3['petal width (cm)'],color='blue',label='blue')
plt.scatter(Km.cluster_centers_[:,0],Km.cluster_centers_[:,1],color='purple',marker='*',label='Centroid',s=90)
plt.legend()
```

```python
km_range = range(1,10)
sse = []

for i in km_range:
    KM = KMeans(n_clusters=i)
    KM.fit(File[['petal length (cm)','petal width (cm)']])
    sse.append(KM.inertia_)
sse
```

```python
Fontx = {'family':'Algerian','color':'blue','size':17}
Fonty = {'family':'century','color':'green','size':17}

plt.xlabel('K Mean',fontdict=Fontx)
plt.ylabel('Sum Of Squared Error',fontdict=Fonty)
plt.title('Elbow Method To find The K.',family='impact',color='Red',size=32)
plt.plot(km_range,sse,linewidth=3)
plt.show()
```
