```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import linear_model

Data = pd.read_csv('D:/Python/File Handling/LogistRegress.csv')
Data
label = LabelEncoder()
Data['Insurance'] = label.fit_transform(Data['Insurance'])
Data
```

```python
X = Data[['Age']]
Y = Data['Insurance']

Model = linear_model.LogisticRegression()

Xtr , Yts , Ytr , Yts = train_test_split(X,Y,test_size=0.2)

Model.fit(Xtr,Ytr)
Model.score(Xtr,Ytr)
Xtr

```

```python
print('0 Indicates No will Buy Insurence')
print('1 Indicates Will Buy Insurence')
Model.predict([[54]])
```
