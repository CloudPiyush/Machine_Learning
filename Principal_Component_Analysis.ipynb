```python
import pandas
from sklearn.datasets import load_digits

Dataset = load_digits()
print(dir(Dataset))
Data = pandas.DataFrame(Dataset.data,columns=Dataset.feature_names)
Data['Target'] = Dataset.target
Data
```

```python
X = Data.drop('Target',axis='columns')
Y = Data['Target']
import sklearn.preprocessing

Scale = sklearn.preprocessing.StandardScaler()
Scale_X = Scale.fit_transform(X)
Scale_X
```

```python
import sklearn.model_selection

Xtrain,Xtest,Ytrain,Ytest =  sklearn.model_selection.train_test_split(Scale_X,Y)
import sklearn.linear_model

Model = sklearn.linear_model.LogisticRegression(max_iter=1000)
Model.fit(Xtrain,Ytrain)
Model.score(Xtest,Ytest)
```

```python
import sklearn.decomposition

Pca = sklearn.decomposition.PCA(.85)
Pca_X = Pca.fit_transform(X)
print('\nThe Shape of Data after reducing dimention using the 85% is :',Pca_X.shape)
print('\nThe Sorted Component are :{0}'.format(Pca.n_components_))
print('\nThe Variennce of the {0} components are :{1}'.format(Pca.n_components_,Pca.explained_variance_ratio_))
```

```python
import sklearn.linear_model 
import sklearn.model_selection
X_train_PCA,X_test_PCA,Ytrain,ytest = sklearn.model_selection.train_test_split(Pca_X,Y,test_size=0.3,random_state=3)
second_model = sklearn.linear_model.LogisticRegression(max_iter=500)
second_model.fit(X_train_PCA,Ytrain)
second_model.score(X_test_PCA,ytest)
```

```python
import sklearn.decomposition
second_Pca = sklearn.decomposition.PCA(n_components=2)
second_Pca_x = second_Pca.fit_transform(X)
second_Pca.n_components_
second_Pca.explained_variance_ratio_
```

```python
import sklearn.linear_model
import sklearn.model_selection
Pca_X_train, Pca_X_test, y_train, y_test = sklearn.model_selection.train_test_split(second_Pca_x,Y,test_size=0.3)

Third_model = sklearn.linear_model.LogisticRegression(solver='liblinear')
Third_model.fit(Pca_X_train,Ytrain)
```

```python
Third_model.score(Pca_X_test,y_test)
```
