```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('D:/Python/File Handling/car_price_encode.csv')
Data
```

```python
encode = LabelEncoder()
Data['Car Model'] = encode.fit_transform(Data['Car Model'])
Data
```

```python
Data.drop('Sell Price($)',axis='columns')
X = Data[['Mileage','Age(yrs)','Car Model']]
```

```python
Y = Data['Sell Price($)']
Y
```

```python
from sklearn import linear_model
Model = linear_model.LinearRegression()
Model.fit(X,Y)
Model.predict([[86000,7,1]])
```

```python
from scipy import stats
A = Data['Age(yrs)']
B = Data['Sell Price($)']
slope,Intercept,r,p,std_err = stats.linregress(A,B)
def Fun(A):
    return slope * A + Intercept
MyModel = list(map(Fun,A))
Age = int(input('\nEnter the Age of car :'))
predict = MyModel.predict(Age)
plt.scatter(A,B)
plt.plot(A,MyModel,marker='P')
plt.show()
```
